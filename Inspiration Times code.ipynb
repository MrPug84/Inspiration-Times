def buscar_campanas_por_palabra_clave(keyword, fields, dataframe):
    """
    Busca campañas en un DataFrame por palabras clave en campos de texto específicos.

    Args:
        keyword (str): La palabra clave a buscar.
        fields (list): Una lista de nombres de columnas (campos de texto) donde buscar.
        dataframe (pd.DataFrame): El DataFrame que contiene los datos de las campañas.

    Returns:
        pd.DataFrame: Un DataFrame con las campañas que coinciden con la palabra clave
                      en al menos uno de los campos especificados.
    """
    if not fields:
        return pd.DataFrame() # Retorna un DataFrame vacío si no hay campos para buscar

    # Convertir la palabra clave a minúsculas para una búsqueda insensible a mayúsculas
    keyword_lower = str(keyword).lower()
    
    # Crear una máscara booleana para filtrar el DataFrame
    mask = pd.Series([False] * len(dataframe)) # Inicializar máscara con Falses

    for field in fields:

            # Usamos .astype(str) para asegurar que el campo sea de tipo string antes de aplicar .str.contains

        else:
            print(f"Advertencia: El campo '{field}' no es válido o no es de tipo string para búsqueda de texto.")

    return dataframe[mask].reset_index(drop=True)

## Implementar Función de Búsqueda por Texto

### Subtask:
Crear una función en Python que permita buscar campañas por palabras clave en los campos de texto como 'nombre_campana', 'descripcion' y 'reconocimientos'. Esta función debería ser flexible para buscar en uno o varios campos.


**Reasoning**:
I need to define the `buscar_campanas_por_palabra_clave` function as requested, which will allow searching campaigns by keywords in specified text fields of the DataFrame. Then, I will demonstrate its usage with the provided examples.



````python
def buscar_campanas_por_palabra_clave(keyword, fields, dataframe=df):
    """
    Busca campañas por palabras clave en los campos de texto especificados.

    Args:
        keyword (str): La palabra clave a buscar.
        fields (list): Una lista de nombres de columnas donde buscar la palabra clave.
        dataframe (pd.DataFrame): El DataFrame donde realizar la búsqueda.

    Returns:
        pd.DataFrame: Un DataFrame con las campañas que coinciden con la búsqueda.
    """
    keyword_lower = keyword.lower()
    # Iniciamos una expresion booleana con todos los valores en False
    # Esto asegura que si no hay campos válidos, la expresion es False por defecto
    mask = pd.Series([False] * len(dataframe), index=dataframe.index)

    for field in fields:
        if field in dataframe.columns and dataframe[field].dtype == 'object':
            # Asegúrate de que la columna se trate como string antes de aplicar .str.contains()
            field_mask = dataframe[field].astype(str).str.contains(keyword_lower, case=False, na=False)
            mask = mask | field_mask
        # Para la columna 'medios' que es una lista de strings
        elif field == 'medios' and field in dataframe.columns:
            # Busca la palabra clave en cualquier elemento de la lista de medios
            field_mask = dataframe[field].apply(lambda x: any(keyword_lower in str(item).lower() for item in x))
            mask = mask | field_mask
        # Para la columna 'tono_comunicacion'
        elif field == 'tono_comunicacion' and field in dataframe.columns and dataframe[field].dtype == 'object':
            field_mask = dataframe[field].astype(str).str.contains(keyword_lower, case=False, na=False)
            mask = mask | field_mask

    return dataframe[mask].reset_index(drop=True)

print("##### Resultados de búsqueda para 'Perú' en 'nombre_campana' y 'descripcion' #####")
resultados_peru = buscar_campanas_por_palabra_clave('Perú', ['nombre_campana', 'descripcion'])
display(resultados_peru[['id', 'nombre_campana', 'descripcion']])

print("##### Resultados de búsqueda para 'Digital' en 'medios' #####")
resultados_digital = buscar_campanas_por_palabra_clave('Digital', ['medios'])
display(resultados_digital[['id', 'nombre_campana', 'medios']])

print("##### Resultados de búsqueda para 'Lápiz' en 'reconocimientos' #####")
resultados_lapiz = buscar_campanas_por_palabra_clave('Lápiz', ['reconocimientos'])
display(resultados_lapiz[['id', 'nombre_campana', 'reconocimientos']])

print("##### Resultados de búsqueda para 'inspirador' en 'tono_comunicacion' #####")
resultados_inspirador = buscar_campanas_por_palabra_clave('inspirador', ['tono_comunicacion'])
display(resultados_inspirador[['id', 'nombre_campana', 'tono_comunicacion']])
````

## Implementar Función de Filtrado por Criterios

### Objetivo:
Desarrollar una función que permita filtrar las campañas basándose en múltiples criterios como 'agencia', 'sector', 'medios', 'tipo_estrategia', 'tono_comunicacion' y 'fecha_lanzamiento'. La función debe poder manejar uno o varios filtros simultáneamente.


````python
def filtrar_campanas_por_criterios(dataframe, criterios):
    """
    Filtra campañas en un DataFrame por múltiples criterios.

    Args:
        dataframe (pd.DataFrame): El DataFrame que contiene los datos de las campañas.
        criterios (dict): Un diccionario donde las claves son los nombres de las columnas
                          y los valores son los criterios de filtrado.
                          - Para columnas de texto (agencia, sector, etc.), el valor es un string.
                          - Para 'medios', el valor es una lista de strings.
                          - Para 'fecha_lanzamiento', el valor es una tupla o lista (fecha_inicio, fecha_fin).

    Returns:
        pd.DataFrame: Un DataFrame con las campañas que cumplen todos los criterios especificados.
    """
    # Iniciamos una expresion booleana con todos los valores en True.
    # Esta se irá actualizando con cada criterio de filtrado.
    mask = pd.Series([True] * len(dataframe), index=dataframe.index)

    for columna, valor_criterio in criterios.items():
        if columna not in dataframe.columns:
            print(f"Advertencia: La columna '{columna}' no existe en el DataFrame. Saltando este criterio.")
            continue

        # Criterios para columnas de texto con coincidencia exacta (insensible a mayúsculas/minúsculas)
        if columna in ['agencia', 'sector', 'tipo_estrategia', 'tono_comunicacion']:
            if pd.api.types.is_string_dtype(dataframe[columna]):
                mask &= (dataframe[columna].str.lower() == str(valor_criterio).lower())
            else:
                print(f"Advertencia: La columna '{columna}' no es de tipo string. Saltando este criterio.")

        # Criterio para la columna 'medios' (lista de strings)
        elif columna == 'medios':
            if isinstance(valor_criterio, list):
                # Creamos una expresion booleana para 'medios'
                medios_sub_mask = pd.Series([False] * len(dataframe), index=dataframe.index)
                for medio_buscado in valor_criterio:
                    # Verificamos si algún medio de la campaña coincide (insensible a mayúsculas/minúsculas)
                    medios_sub_mask |= dataframe['medios'].apply(lambda x:
                        any(str(medio_buscado).lower() == str(m).lower() for m in x)
                    )
                mask &= medios_sub_mask
            else:
                print(f"Advertencia: El criterio para 'medios' debe ser una lista. Saltando este criterio.")

        # Criterio para la columna 'fecha_lanzamiento' (rango de fechas)
        elif columna == 'fecha_lanzamiento':
            if isinstance(valor_criterio, (tuple, list)) and len(valor_criterio) == 2:
                try:
                    fecha_inicio = pd.to_datetime(valor_criterio[0])
                    fecha_fin = pd.to_datetime(valor_criterio[1])
                    mask &= (dataframe['fecha_lanzamiento'] >= fecha_inicio) & \
                            (dataframe['fecha_lanzamiento'] <= fecha_fin)
                except ValueError:
                    print(f"Advertencia: Las fechas en el criterio para 'fecha_lanzamiento' no son válidas. Saltando este criterio.")
            else:
                print(f"Advertencia: El criterio para 'fecha_lanzamiento' debe ser una tupla/lista (fecha_inicio, fecha_fin). Saltando este criterio.")
        else:
            print(f"Advertencia: Criterio no manejado para la columna '{columna}'. Saltando este criterio.")

    # Aplicamos la expresion booleana final al DataFrame y devuelve el resultado.
    return dataframe[mask].reset_index(drop=True)



print("##### Ejemplo 1: Filtrar por una sola columna de texto (agencia='McCann') #####")
criterios_1 = {'agencia': 'McCann'}
resultados_1 = filtrar_campanas_por_criterios(df, criterios_1)
display(resultados_1[['id', 'nombre_campana', 'agencia']])

print("\n##### Ejemplo 2: Filtrar por múltiples columnas de texto (sector='Finanzas', tono_comunicacion='Inspirador') #####")
criterios_2 = {'sector': 'Finanzas', 'tono_comunicacion': 'Inspirador'}
resultados_2 = filtrar_campanas_por_criterios(df, criterios_2)
display(resultados_2[['id', 'nombre_campana', 'sector', 'tono_comunicacion']])

print("\n##### Ejemplo 3: Combinación de texto, medios y rango de fechas (agencia='VML', medios=['Digital', 'OOH'], fecha_lanzamiento=('2023-01-01', '2024-12-31')) #####")
criterios_3 = {
    'agencia': 'VML',
    'medios': ['Digital', 'OOH'],
    'fecha_lanzamiento': ('2023-01-01', '2024-12-31')
}
resultados_3 = filtrar_campanas_por_criterios(df, criterios_3)
display(resultados_3[['id', 'nombre_campana', 'agencia', 'medios', 'fecha_lanzamiento']])
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

1# Importamos librerias

# 2. Visualización 1: Distribución de campañas por agencia.
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='agencia', order=df['agencia'].value_counts().index, palette='viridis')
plt.title('Distribución de Campañas por Agencia')
plt.xlabel('Agencia')
plt.ylabel('Número de Campañas')
plt.xticks(rotation=45, ha='right') # Rotamos etiquetas para mejor legibilidad
plt.tight_layout()
plt.show()

# 3. Visualización 2: Distribución de campañas por sector.
plt.figure(figsize=(10, 8))
df['sector'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='plasma')
plt.title('Distribución de Campañas por Sector')
plt.ylabel('') # Ocultamos la etiqueta 'sector' en el eje y para un gráfico de pastel
plt.tight_layout()
plt.show()

# 4. Visualización 3: Distribución de campañas por tipo de medio.
# Desapilar los medios para contar cada uno individualmente
medios_desapilados = df['medios'].explode()
plt.figure(figsize=(12, 6))
sns.countplot(data=None, x=medios_desapilados, order=medios_desapilados.value_counts().index, palette='magma')
plt.title('Distribución de Campañas por Tipo de Medio')
plt.xlabel('Tipo de Medio')
plt.ylabel('Número de Campañas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 5. Visualización 4: Evolución de 'tipo_estrategia' a lo largo del tiempo.
# Extraemos el año de la fecha de lanzamiento
df['año_lanzamiento'] = df['fecha_lanzamiento'].dt.year

# Seleccionamos las 5 estrategias más comunes para evitar un gráfico saturado
estrategias_comunes = df['tipo_estrategia'].value_counts().nlargest(5).index
df_filtrado_estrategias = df[df['tipo_estrategia'].isin(estrategias_comunes)]

# Agrupamos por año y tipo de estrategia y contar campañas
tendencia_estrategias = df_filtrado_estrategias.groupby(['año_lanzamiento', 'tipo_estrategia']).size().unstack(fill_value=0)

plt.figure(figsize=(14, 7))
tendencia_estrategias.plot(kind='line', marker='o', ax=plt.gca())
plt.title('Tendencia de Tipos de Estrategia a lo largo del Tiempo (Top 5)')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Número de Campañas')
plt.legend(title='Tipo de Estrategia', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

````python
import streamlit as st
import pandas as pd
import io
import matplotlib.pyplot as plt
import seaborn as sns

# --- Definición de funciones de búsqueda y filtrado ---

def buscar_campanas_por_palabra_clave(keyword, fields, dataframe):
    """
    Busca campañas por palabras clave en los campos de texto especificados.

    Args:
        keyword (str): La palabra clave a buscar.
        fields (list): Una lista de nombres de columnas donde buscar la palabra clave.
        dataframe (pd.DataFrame): El DataFrame donde realizar la búsqueda.

    Returns:
        pd.DataFrame: Un DataFrame con las campañas que coinciden con la búsqueda.
    """
    keyword_lower = keyword.lower()
    # Iniciamos una expresión booleana con todos los valores en False
    # Esto asegura que si no hay campos válidos, la expresión es False por defecto
    mask = pd.Series([False] * len(dataframe), index=dataframe.index)

    for field in fields:
        if field in dataframe.columns and dataframe[field].dtype == 'object':
            # Asegúrate de que la columna se trate como string antes de aplicar .str.contains()
            field_mask = dataframe[field].astype(str).str.contains(keyword_lower, case=False, na=False)
            mask = mask | field_mask
        # Para la columna 'medios' que es una lista de strings
        elif field == 'medios' and field in dataframe.columns:
            # Busca la palabra clave en cualquier elemento de la lista de medios
            field_mask = dataframe[field].apply(lambda x: any(keyword_lower in str(item).lower() for item in x))
            mask = mask | field_mask
        # Para la columna 'tono_comunicacion' (ya está cubierta por el primer if si es 'object')
        # Si necesitara un manejo especial, se añadiría aquí.

    return dataframe[mask].reset_index(drop=True)

def filtrar_campanas_por_criterios(dataframe, criterios):
    """
    Filtra campañas en un DataFrame por múltiples criterios.

    Args:
        dataframe (pd.DataFrame): El DataFrame que contiene los datos de las campañas.
        criterios (dict): Un diccionario donde las claves son los nombres de las columnas
                          y los valores son los criterios de filtrado.
                          - Para columnas de texto (agencia, sector, etc.), el valor es un string.
                          - Para 'medios', el valor es una lista de strings.
                          - Para 'fecha_lanzamiento', el valor es una tupla o lista (fecha_inicio, fecha_fin).

    Returns:
        pd.DataFrame: Un DataFrame con las campañas que cumplen todos los criterios especificados.
    """
    # Iniciamos una expresión booleana con todos los valores en True.
    # Esta se irá actualizando con cada criterio de filtrado.
    mask = pd.Series([True] * len(dataframe), index=dataframe.index)

    for columna, valor_criterio in criterios.items():
        if columna not in dataframe.columns:
            # st.warning(f"Advertencia: La columna '{columna}' no existe en el DataFrame. Saltando este criterio.")
            continue

        # Criterios para columnas de texto con coincidencia exacta (insensible a mayúsculas/minúsculas)
        if columna in ['agencia', 'sector', 'tipo_estrategia', 'tono_comunicacion']:
            if pd.api.types.is_string_dtype(dataframe[columna]):
                mask &= (dataframe[columna].str.lower() == str(valor_criterio).lower())
            else:
                # st.warning(f"Advertencia: La columna '{columna}' no es de tipo string. Saltando este criterio.")
                pass

        # Criterio para la columna 'medios' (lista de strings)
        elif columna == 'medios':
            if isinstance(valor_criterio, list):
                # Creamos una expresión booleana para 'medios'
                medios_sub_mask = pd.Series([False] * len(dataframe), index=dataframe.index)
                for medio_buscado in valor_criterio:
                    # Verificamos si algún medio de la campaña coincide (insensible a mayúsculas/minúsculas)
                    medios_sub_mask |= dataframe['medios'].apply(lambda x:
                        any(str(medio_buscado).lower() == str(m).lower() for m in x)
                    )
                mask &= medios_sub_mask
            else:
                # st.warning(f"Advertencia: El criterio para 'medios' debe ser una lista. Saltando este criterio.")
                pass

        # Criterio para la columna 'fecha_lanzamiento' (rango de fechas)
        elif columna == 'fecha_lanzamiento':
            if isinstance(valor_criterio, (tuple, list)) and len(valor_criterio) == 2:
                try:
                    fecha_inicio = pd.to_datetime(valor_criterio[0])
                    fecha_fin = pd.to_datetime(valor_criterio[1])
                    mask &= (dataframe['fecha_lanzamiento'] >= fecha_inicio) & \
                            (dataframe['fecha_lanzamiento'] <= fecha_fin)
                except ValueError:
                    # st.warning(f"Advertencia: Las fechas en el criterio para 'fecha_lanzamiento' no son válidas. Saltando este criterio.")
                    pass
            else:
                # st.warning(f"Advertencia: El criterio para 'fecha_lanzamiento' debe ser una tupla/lista (fecha_inicio, fecha_fin). Saltando este criterio.")
                pass
        # else:
            # st.warning(f"Advertencia: Criterio no manejado para la columna '{columna}'. Saltando este criterio.")

    # Aplicamos la expresión booleana final al DataFrame y devuelve el resultado.
    return dataframe[mask].reset_index(drop=True)

# --- Cargar y Preprocesar Datos ---

datos_string_csv = """id,nombre_campana,marca,agencia,sector,medios,tipo_estrategia,fecha_lanzamiento,tono_comunicacion,reconocimientos,descripcion
1,Perú Eres Tú,Perú (Marca País),McCann,Turismo,TV|Radio|Digital,Identidad Nacional,2023-01-15,Emocional,Festival Cannes,Campaña que posiciona Perú como destino turístico
2,Ama lo Tuyo,Inca Kola,VML,Bebidas,TV|OOH|Digital,Nostalgia y Orgullo,2022-06-20,Desenfadado,Lápiz de Oro,Reposicionamiento basado en patrimonio cultural
3,La Voz del Pueblo,Movistar,Circus Grey,Telecomunicaciones,Digital|Social Media,Engagement Social,2024-03-10,Inspirador,Effie Awards,Plataforma de participación ciudadana
4,Juntos Somos Más,BCP,Fahrenheit DDB,Finanzas,TV|Digital|Experiencial,Inclusión Financiera,2023-09-05,Humano,IAB Peru Awards,Campaña sobre educación financiera
5,Destapa lo Mejor,Corona,Digitas,Bebidas,Digital|Influencers|OOH,Experiencia,2024-01-12,Joven,Cannes Lions,Activación con micro-influencers
6,Poder Femenino,Avon,121 Latam,Belleza,Digital|TV|Social,Empoderamiento,2023-05-22,Inspirador,Premios Óscar de Publicidad,Celebra mujeres emprendedoras
7,Sin Límites,Nike,Valor,Deporte,Digital|OOH|Experiencias,Rendimiento,2024-02-14,Motivacional,Festival Publicidad Lima,Atletas peruanos como protagonistas
8,Recuerda Quién Eres,Telefónica,Ogilvy Perú,Telecomunicaciones,TV|Digital,Identidad,2023-11-08,Nostálgico,Premio Aníbal Ford,Conexión emocional con usuarios
9,Reinvéntate,Scotiabank,Boost Brand,Finanzas,Digital|Redes Sociales,Transformación,2024-04-19,Optimista,Viral en Redes,Adaptación a nuevas economías
10,Sabor a Perú,Gloria,McCann,Alimentos,TV|OOH|Digital,Patrimonio Culinario,2023-07-30,Cálido,Lápiz de Oro,Productos lácteos con raíces peruanas
11,Somos Resilientes,Agua Oxigenada Mercononi,VML,Salud,Digital|Influencers,Resiliencia,2024-05-10,Humano,Trending Topic,Mensajes positivos post-pandemia
12,La Conexión Verdadera,Claro,Circus Grey,Telecomunicaciones,TV|Digital|Social,Comunidad,2023-08-15,Emocional,Premios Lima Advertising,Humanos conectados
13,Tu Futuro Empieza Hoy,Beca Perú,Fahrenheit DDB,Educación,Digital|TV,Aspiración,2024-06-01,Inspirador,IAB Awards,Becas para jóvenes peruanos
14,Muévete,Reebok,Digitas,Deporte,Digital|Influencers,Actividad Física,2023-10-25,Energético,Festival Publicidad,Movimiento y bienestar
15,Raíces Profundas,Quilmes,121 Latam,Bebidas,OOH|TV|Digital,Tradición,2024-03-20,Nostálgico,Cannes Lions,Herencia familiar
16,Vive Más,Seguros Integra,Valor,Seguros,Digital|TV,Calidad de Vida,2023-12-05,Tranquilizador,Lápiz de Oro,Protección familiar
17,El Poder está en Ti,Nestlé,Ogilvy Perú,Alimentos,TV|Digital|Experiencial,Empoderamiento,2024-02-28,Tono Motivacional,Premio Aníbal Ford,Nutrición y desarrollo
18,Más Cerca,Interbank,Boost Brand,Finanzas,Digital|Social,Proximidad,2024-01-30,Cálido,IAB Awards,Banking digital accesible
19,Herencia de Innovación,Samsung,McCann,Tecnología,Digital|TV|OOH,Innovación,2023-09-12,Futurista,Cannes Lions,Tecnología con raíces peruanas
20,Juntos por Perú,Caja Arequipa,VML,Finanzas,Digital|Experiencial,Solidaridad,2024-04-05,Inspirador,Festival Publicidad,Microfinanzas solidarias"""

df = pd.read_csv(io.StringIO(datos_string_csv), sep=',')
df['medios'] = df['medios'].apply(lambda x: x.split('|'))
df['fecha_lanzamiento'] = pd.to_datetime(df['fecha_lanzamiento'])
df['año_lanzamiento'] = df['fecha_lanzamiento'].dt.year

# Traducción de los valores de la columna 'medios'
media_translation_map = {
    'TV': 'Televisión',
    'Radio': 'Radio',
    'Digital': 'Digital',
    'OOH': 'Publicidad Exterior',
    'Influencers': 'Influencers',
    'Social Media': 'Redes Sociales',
    'Social': 'Redes Sociales',
    'Experiencial': 'Experiencial',
    'Experiencias': 'Experiencial' # Consolidar 'Experiencias' en 'Experiencial'
}

df['medios'] = df['medios'].apply(lambda x: [media_translation_map.get(medio, medio) for medio in x])


# --- Configuración de la interfaz de usuario Streamlit ---

st.title('Inspira Times')

# Campo de entrada de texto para búsqueda de palabras clave
keyword_search = st.text_input('Buscar campaña por palabra clave...', '')

# Barra lateral para filtros
st.sidebar.header('Filtros de Campañas')

# Filtro por 'Agencia'
agencias_unicas = ['Todas'] + sorted(df['agencia'].unique().tolist())
agencia_filter = st.sidebar.selectbox('Filtrar por Agencia', agencias_unicas)

# Filtro por 'Sector'
sectores_unicos = ['Todos'] + sorted(df['sector'].unique().tolist())
sector_filter = st.sidebar.selectbox('Filtrar por Sector', sectores_unicos)

# Filtro por 'Medios'
medios_unicos = sorted(list(set(item for sublist in df['medios'] for item in sublist)))
medios_filter = st.sidebar.multiselect('Filtrar por Medios', medios_unicos)

# Filtro por 'fecha_lanzamiento'
st.sidebar.subheader('Rango de Fechas de Lanzamiento')
min_date_df = df['fecha_lanzamiento'].min().to_pydatetime().date()
max_date_df = df['fecha_lanzamiento'].max().to_pydatetime().date()
start_date = st.sidebar.date_input('Fecha de Inicio', value=min_date_df)
end_date = st.sidebar.date_input('Fecha de Fin', value=max_date_df)


# --- Lógica de Búsqueda y Filtrado ---

# Inicializa una variable `df_actual` con el DataFrame `df` original.
df_actual = df.copy()

# Si `keyword_search` no está vacío, aplica la función `buscar_campanas_por_palabra_clave`.
if keyword_search:
    df_actual = buscar_campanas_por_palabra_clave(
        keyword_search,
        ['nombre_campana', 'descripcion', 'reconocimientos', 'tono_comunicacion'],
        df_actual
    )

# Crea un diccionario vacío llamado `criterios_filtrado`.
criterios_filtrado = {}

# Si `agencia_filter` no es 'Todas', añade este criterio.
if agencia_filter != 'Todas':
    criterios_filtrado['agencia'] = agencia_filter

# Si `sector_filter` no es 'Todos', añade este criterio.
if sector_filter != 'Todos':
    criterios_filtrado['sector'] = sector_filter

# Si `medios_filter` no está vacío, añade esta lista de medios.
if medios_filter:
    criterios_filtrado['medios'] = medios_filter

# Si la `start_date` es anterior o igual a la `end_date`, añade la tupla de fechas.
if start_date <= end_date:
    criterios_filtrado['fecha_lanzamiento'] = (start_date, end_date)
else:
    st.sidebar.error('La fecha de inicio no puede ser posterior a la fecha de fin.')

# Si el diccionario `criterios_filtrado` no está vacío, aplica la función `filtrar_campanas_por_criterios`.
if criterios_filtrado:
    df_actual = filtrar_campanas_por_criterios(df_actual, criterios_filtrado)

# --- Resultados de las Campañas ---
st.subheader('Resultados de las Campañas')
if not df_actual.empty:
    st.dataframe(df_actual)
else:
    st.write("No se encontraron campañas con los criterios seleccionados.")

# --- Visualizaciones de Tendencias ---
st.subheader('Visualizaciones de Tendencias')

if not df_actual.empty:
    # Visualización 1: Distribución de campañas por agencia.
    fig1, ax1 = plt.subplots(figsize=(12, 6))
    sns.countplot(data=df_actual, x='agencia', order=df_actual['agencia'].value_counts().index, palette='viridis', hue='agencia', legend=False, ax=ax1)
    ax1.set_title('Distribución de Campañas por Agencia')
    ax1.set_xlabel('Agencia')
    ax1.set_ylabel('Número de Campañas')
    ax1.tick_params(axis='x', rotation=45) # ha='right' eliminado
    plt.tight_layout()
    st.pyplot(fig1)

    # Visualización 2: Distribución de campañas por sector.
    fig2, ax2 = plt.subplots(figsize=(10, 8))
    df_actual['sector'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='plasma', ax=ax2)
    ax2.set_title('Distribución de Campañas por Sector')
    ax2.set_ylabel('')
    plt.tight_layout()
    st.pyplot(fig2)

    # Visualización 3: Distribución de campañas por tipo de medio.
    medios_desapilados_filtered = df_actual['medios'].explode()
    if not medios_desapilados_filtered.empty:
        fig3, ax3 = plt.subplots(figsize=(12, 6))
        sns.countplot(x=medios_desapilados_filtered, order=medios_desapilados_filtered.value_counts().index, palette='magma', hue=medios_desapilados_filtered, legend=False, ax=ax3)
        ax3.set_title('Distribución de Campañas por Tipo de Medio')
        ax3.set_xlabel('Tipo de Medio')
        ax3.set_ylabel('Número de Campañas')
        ax3.tick_params(axis='x', rotation=45) # ha='right' eliminado
        plt.tight_layout()
        st.pyplot(fig3)
    else:
        st.write("No hay datos de medios para mostrar visualizaciones.")

    # Visualización 4: Evolución de 'tipo_estrategia' a lo largo del tiempo.
    # Extraemos el año de la fecha de lanzamiento (ya está en df_actual por el preprocesamiento)

    # Seleccionamos las 5 estrategias más comunes para evitar un gráfico saturado
    estrategias_comunes_filtered = df_actual['tipo_estrategia'].value_counts().nlargest(5).index
    df_filtrado_estrategias_viz = df_actual[df_actual['tipo_estrategia'].isin(estrategias_comunes_filtered)]

    if not df_filtrado_estrategias_viz.empty:
        # Agrupamos por año y tipo de estrategia y contar campañas
        tendencia_estrategias_filtered = df_filtrado_estrategias_viz.groupby(['año_lanzamiento', 'tipo_estrategia']).size().unstack(fill_value=0)

        fig4, ax4 = plt.subplots(figsize=(14, 7))
        tendencia_estrategias_filtered.plot(kind='line', marker='o', ax=ax4)
        ax4.set_title('Tendencia de Tipos de Estrategia a lo largo del Tiempo (Top 5)')
        ax4.set_xlabel('Año de Lanzamiento')
        ax4.set_ylabel('Número de Campañas')
        ax4.legend(title='Tipo de Estrategia', bbox_to_anchor=(1.05, 1), loc='upper left')
        ax4.grid(True, linestyle='--', alpha=0.7)
        plt.tight_layout()
        st.pyplot(fig4)
    else:
        st.write("No hay datos de estrategias comunes para mostrar tendencias.")
else:
    st.write("No hay datos disponibles para generar visualizaciones.")
````
